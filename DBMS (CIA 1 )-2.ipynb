```python
conn = sqlite3.connect("financial_data.db")
df.to_sql("currency_rates", conn, if_exists="replace", index=False)
```

```python
pd.read_sql("SELECT * FROM currency_rates",conn)
```

```python
#WHERE
pd.read_sql("SELECT * FROM currency_rates WHERE [Currency] is 'AED'",conn)
```

```python
#GROUP BY
#Groups results by currency and counts how many rows per currency.
pd.read_sql("""SELECT Currency, COUNT(*) AS currency_count
FROM currency_rates
GROUP BY Currency""",conn)
```

```python
#ORDER BY
#Orders currencies by their value in USD, from highest to lowest.
pd.read_sql("SELECT * FROM currency_rates ORDER BY [USD per unit] DESC",conn)
```

```python
#LIMIT
#Returns the top 5 most valuable currencies per USD.
pd.read_sql("SELECT * FROM currency_rates ORDER BY [USD per unit] DESC LIMIT 5",conn)
```

```python
#AS
#Renames the columns in the result set.
pd.read_sql("SELECT Currency AS Code, Name AS Country_Name, [USD per unit] AS Value_in_USD FROM currency_rates",conn)
```

```python
#HAVING
#Filters grouped data where a currency appears more than once.
pd.read_sql("""SELECT Currency, COUNT(*) AS cnt
FROM currency_rates
GROUP BY Currency
HAVING COUNT(*) > 1""",conn)
```

```python
#BETWEEN AND
#Returns currencies where 1 unit is worth between 0.5 and 1.5 USD.
pd.read_sql("""SELECT * FROM currency_rates
WHERE [USD per unit] BETWEEN 0.5 AND 1.5""",conn)
```

```python
#AVG, SUM, COUNT
#Calculates average, sum, and count of USD per unit.
pd.read_sql("""
    SELECT 
      AVG("USD per unit") AS avg_usd,
      SUM("USD per unit") AS total_usd,
      COUNT(*) AS total_currencies
    FROM currency_rates
""", conn)
```

```python
#DISTINCT
pd.read_sql("SELECT DISTINCT Currency FROM currency_rates",conn)
```

```python
#UPDATE
conn.execute("""
UPDATE currency_rates
SET [USD per unit] = 1
WHERE Currency = 'USD'
""")
conn.commit()  
```

```python
#INSERT
conn.execute("""INSERT INTO currency_rates (Currency, Name, [Units per USD],[USD per unit])
VALUES ('XYZ', 'Fantasyland Dollar', 1000, 0.001)""")
pd.read_sql("SELECT * FROM currency_rates WHERE Currency is 'XYZ'",conn)
```

```python
#LIKE
pd.read_sql("""SELECT * FROM currency_rates
             WHERE Name LIKE '%Dollar%'""",conn)

```
